```python
import pandas as pd
```

```python
train_pos_neg_file = "./../data/train_data/MSCG/70k_train_ms_concept_graph.tsv"
valid_pos_neg_file = "./../data/train_data//MSCG/70k_valid_ms_concept_graph.tsv"
test_pos_neg_file = "./../data/evaluation_data/MSCG/70k_test_ms_concept_graph.tsv"
```

```python
df_train_pos_neg = pd.read_csv(train_pos_neg_file, sep="\t", header=None, names = ["concept", "property", "label"])
```

```python
df_train_pos_neg
```

```python
df_valid_pos_neg = pd.read_csv(valid_pos_neg_file, sep="\t", header=None, names = ["concept", "property", "label"])
```

```python
df_valid_pos_neg
```

```python
df_test_pos_neg = pd.read_csv(test_pos_neg_file, sep="\t", header=None, names = ["concept", "property", "label"])
```

```python
df_test_pos_neg
```

```python
df_pos_train = df_train_pos_neg[df_train_pos_neg["label"] == 1]
```

```python
df_pos_train
```

```python
# Checking overlap with test data
df_overlap_train_test = df_test_pos_neg.merge(df_pos_train, how = 'inner', on = ["concept", "property"], indicator=False)
df_overlap_train_test
```

```python
df_pos_valid = df_valid_pos_neg[df_valid_pos_neg["label"] == 1]
df_pos_valid
```

```python
# Checking overlap with test data
df_overlap_valid_test = df_test_pos_neg.merge(df_pos_valid, how = 'inner', on = ["concept", "property"], indicator=False)
df_overlap_valid_test
```

```python
df_pos_train = df_pos_train.drop("label", axis=1)
```

```python
df_pos_train.to_csv("./../data/train_data/MSCG/70k_nbs_pos_train_mscg.tsv", sep="\t", index=None, header=None)
```

```python
df_pos_valid = df_pos_valid.drop("label", axis=1)
```

```python
df_pos_valid.to_csv("./../data/train_data/MSCG/70k_nbs_pos_valid_mscg.tsv", sep="\t", index=None, header=None)
```
