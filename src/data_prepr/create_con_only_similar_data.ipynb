```python
import pandas as pd
import os
```

```python
def create_con_only_similar_data(input_file, con_similar_file, data_type=None):
    
    inp_file_path, inp_file_name = os.path.split(input_file)
    
    print (f"Input File Path : {inp_file_path}")
    print (f"Input FIle Name : {inp_file_name}")
    
    inp_df = pd.read_csv(input_file, sep="\t", names=["concept",  "predict_property", "label"])[0:100]
    
    inp_concepts = inp_df["concept"].unique() 
    num_inp_concepts = len(inp_concepts)
    
    print (f"Input Data DF")
    print (inp_df)
    print (f"num_inp_concepts : {num_inp_concepts}")
    print ()
        
    con_similar_data = pd.read_csv(con_similar_file, sep="\t", names=["concept",  "similar_property"])
    con_similar_concepts = con_similar_data["concept"].unique()
    num_con_similar_concepts = len(con_similar_concepts)
    
    print ()
    print (f"Concept Similar Property File")
    print (con_similar_data)
    print (f"num_con_similar_concepts : {num_con_similar_concepts}")
    
    all_prop_augmented_data = []
    for idx, (concept, predict_property, label) in enumerate(zip(inp_df["concept"], inp_df["predict_property"], inp_df["label"])):
        
        print (f"concept, predict_property, label : {concept, predict_property, label}")
        similar_props = con_similar_data[con_similar_data["concept"] == concept]["similar_property"].unique().tolist()
        
        similar_props = ", ".join(similar_props)
        print (f"similar_props : {similar_props}")
        
        all_prop_augmented_data.append((concept, similar_props, similar_props, predict_property, label))
        
        print (f"Data : {(concept, similar_props, similar_props, predict_property, label)}")
        print ()
    
    
    df = pd.DataFrame.from_records(all_prop_augmented_data)
    
    save_file_name = os.path.join(inp_file_path, f"con_only_similar_data_{inp_file_name}")
    
    df.to_csv(save_file_name, sep="\t", index=None, header=None)
    
    print (f"Only Concept Augmented Data Saved in  {save_file_name}")
        
input_file = "./../../data/train_data/je_con_prop/5_neg_train_cnet_premium.tsv"
con_similar_file = "./../../data/train_data/cnetp_con_similar_5_prop_vocab.tsv"


create_con_only_similar_data(input_file=input_file, con_similar_file=con_similar_file)
    
```
